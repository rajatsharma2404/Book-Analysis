# Load the book

```python
with open("miracle_in_the_andes.txt", encoding = "utf-8") as file:
    book = file.read()
```

# Count the words

```python
import re
pattern = re.compile("[a-zA-Z]+")
findings = re.findall(pattern, book)
dict = {}

for word in findings:
    if word.lower() in dict.keys():
        dict[word.lower()] = dict[word.lower()]+1
    else:
        dict[word.lower()] = 1

dict_l = [(count, word) for (word, count) in dict.items()]
sorted(dict_l, reverse=True)
"Hello"
```

```python
import nltk
from nltk.corpus import stopwords
nltk.download("stopwords")
```

```python
english_stopwords = stopwords.words("english")

nonstop_list = []
for count, word in dict_l:
    if word not in english_stopwords:
        nonstop_list.append((count, word))

sorted(nonstop_list, reverse=True)
        
```

# Sentiment Analysis

```python
from nltk.sentiment import SentimentIntensityAnalyzer
```

```python
nltk.download('vader_lexicon')
analyzer = SentimentIntensityAnalyzer()
```

```python
scores = analyzer.polarity_scores("Hello, It is a beautiful day")
scores
```

### Sentiment Analysis for each chapter

```python
pattern = re.compile("Chapter [0-9]+")
findings = re.split(pattern, book)
chapters = findings[1:]
len(chapters)
```

```python
sentiment_dict = []
for chapter in chapters:
    score = analyzer.polarity_scores(chapter)
    sentiment_dict.append(score)

sentiment_dict
```
